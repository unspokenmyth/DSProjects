```python
atp['Year'] = pd.to_datetime(atp['tourney_date'], format='%Y%m%d').dt.year
```

```python
years = atp['Year'].unique()
yearly_stats = pd.DataFrame(years, columns=['Year'])

# Step 5: Calculate Federer's wins and total matches per year
federer_wins = []
federer_total_matches = []

for year in years:
    federer_matches = atp[atp['Year'] == year]
    federer_wins_count = federer_matches[federer_matches['winner_name'] == 'Roger Federer'].shape[0]
    federer_total_count = federer_matches[(federer_matches['winner_name'] == 'Roger Federer') | 
                                          (federer_matches['loser_name'] == 'Roger Federer')].shape[0]
    
    federer_wins.append(federer_wins_count)
    federer_total_matches.append(federer_total_count)

yearly_stats['Federer_Wins'] = federer_wins
yearly_stats['Federer_Total_Matches'] = federer_total_matches

# Step 6: Calculate Nadal's wins and total matches per year
nadal_wins = []
nadal_total_matches = []

for year in years:
    nadal_matches = atp[atp['Year'] == year]
    nadal_wins_count = nadal_matches[nadal_matches['winner_name'] == 'Rafael Nadal'].shape[0]
    nadal_total_count = nadal_matches[(nadal_matches['winner_name'] == 'Rafael Nadal') | 
                                      (nadal_matches['loser_name'] == 'Rafael Nadal')].shape[0]
    
    nadal_wins.append(nadal_wins_count)
    nadal_total_matches.append(nadal_total_count)

yearly_stats['Nadal_Wins'] = nadal_wins
yearly_stats['Nadal_Total_Matches'] = nadal_total_matches

# Step 7: Calculate win percentages for both players
yearly_stats['Federer_Win_Percentage'] = (yearly_stats['Federer_Wins'] / yearly_stats['Federer_Total_Matches']) * 100
yearly_stats['Nadal_Win_Percentage'] = (yearly_stats['Nadal_Wins'] / yearly_stats['Nadal_Total_Matches']) * 100

# Handle division by zero if there are no matches played
yearly_stats.fillna(0, inplace=True)
```

```python
yearly_stats
```

```python
yearly_stats_filtered = yearly_stats[yearly_stats['Year'] >= 1998]

plt.figure(figsize=(10, 6))
plt.plot(yearly_stats_filtered['Year'], yearly_stats_filtered['Federer_Win_Percentage'], label='Federer', color='#32de84')  # A nice blue
plt.plot(yearly_stats_filtered['Year'], yearly_stats_filtered['Nadal_Win_Percentage'], label='Nadal', color='#ff6347')    # Tomato red
plt.title('Win Percentage of Federer and Nadal Over The Years', fontsize=14, color='black')
plt.xlabel('Year', fontsize=12, color='gray')
plt.ylabel('Win Percentage', fontsize=12, color='gray')
plt.legend(facecolor='lightgrey', framealpha=0.5)
plt.grid(True, color='grey', linestyle='--', linewidth=0.5)
plt.ylim(bottom = 40)
plt.show()
```
