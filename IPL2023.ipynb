```python
def concatenate_outcomes(series):
    return list(series)

bowlerResult = ball.groupby(['bowler'])['outcome'].agg(concatenate_outcomes).reset_index()
bowlerResult = bowlerResult.rename(columns={'outcome': 'sum_outcomes'})
bowlerResult
```

```python
def calculate_bowler_points(bowler_stats):
    scoring_rules = {
        '0': 5,'1lb': 1,'1': 1,'4': -2,'w': 10,
        '6': -6,'1nb': -1,'4lb': -4,'2': 2,'1b': 1,
        '1wd': 1,'2nb': -2,'5nb': -5,'7nb': -7,'3': 3,
        '5wd': -5,'4b': -4,'2wd': 2,'5': -5,'3wd': -3,
        '2lb': -2,'3nb': -3,'2b': -2,
    }

    total_points = 0
    total_runs_conceded = 0
    total_wickets_taken = 0
    total_balls_bowled = len(bowler_stats) 

    for stat in bowler_stats:
        if stat in scoring_rules:
            total_points += scoring_rules[stat]
        if stat.isdigit(): 
            total_runs_conceded += int(stat)
        elif stat == 'w':
            total_wickets_taken += 1

    bowling_average = total_runs_conceded / total_wickets_taken if total_wickets_taken > 0 else 0
    economy_rate = (total_runs_conceded / total_balls_bowled) * 6 

    bowling_average_threshold = 25
    economy_rate_threshold = 6.5

    if bowling_average <= bowling_average_threshold:
        total_points += 20
    if economy_rate <= economy_rate_threshold:
        total_points += 15

    return total_points
```

```python
points_list = []
for bowler_stats in bowlerResult.sum_outcomes:
    points = calculate_bowler_points(bowler_stats)
    points_list.append(points)

bowlerResult['Points'] = points_list
bowlerResult = bowlerResult.sort_values(by = 'Points',ascending = False)
bowlerResult.head(10)
```

```python
bowler_names = bowlerResult.head(10).bowler.tolist()
bowler_points = bowlerResult.head(10).Points.tolist()
plt.figure(figsize=(10, 6))

bars = plt.bar(bowler_names, bowler_points, color='lightgreen')
for bar, points in zip(bars, bowler_points):
    plt.text(bar.get_x() + bar.get_width() / 4 - 0.1, bar.get_height() + 5, str(points), fontsize=10)
plt.xlabel("Bowler Names")
plt.ylabel("Points")
plt.title("Bowler Points")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
batterResult = ball.groupby(['batter'])['outcome'].agg(concatenate_outcomes).reset_index()
batterResult = batterResult.rename(columns={'outcome': 'sum_outcomes'})
batterResult
```

```python
def calculate_batter_points(batter_stats):
    scoring_rules = {
        '0': 0, '1': 1, '2': 2, '3': 3, '4': 6,       
        '6': 10, 'w': -5, '1lb': 0, '1b': 0, '1nb': 0,
        '1wd': 0, '2lb': 0, '2b': 0, '3lb': 0, '3b': 0       
    }

    total_points = 0
    total_runs_scored = 0
    total_balls_faced = len(batter_stats)
    total_boundaries = 0
    total_sixes = 0
    total_dismissals = 0

    for stat in batter_stats:
        if stat in scoring_rules:
            total_points += scoring_rules[stat]
        if stat.isdigit():
            total_runs_scored += int(stat)
            if int(stat) == 4:
                total_boundaries += 1
            elif int(stat) == 6:
                total_sixes += 1
        elif stat == 'w':
            total_dismissals += 1

    strike_rate = (total_runs_scored / total_balls_faced) * 100 if total_balls_faced > 0 else 0

    strike_rate_bonus_threshold = 150.0 
    if strike_rate >= strike_rate_bonus_threshold:
        total_points += 10

    if total_runs_scored == 0 and total_dismissals > 0:
        total_points -= 10

    boundary_bonus_threshold = 5
    if total_boundaries >= boundary_bonus_threshold:
        total_points += 10

    six_bonus_threshold = 3
    if total_sixes >= six_bonus_threshold:
        total_points += 15

    return total_points
```

```python
points_list = []
for batter_stats in batterResult.sum_outcomes:
    points = calculate_batter_points(batter_stats)
    points_list.append(points)

batterResult['Points'] = points_list
batterResult = batterResult.sort_values(by = 'Points',ascending = False)
batterResult.head(10)
```

```python
batter_names = batterResult.head(10).batter.tolist()
batter_points = batterResult.head(10).Points.tolist()
plt.figure(figsize=(10, 6))

bars = plt.bar(batter_names, batter_points, color='lightgreen')

plt.xlabel("Batsmen")
plt.ylabel("Points")
plt.title("Batsmen Points")
for bar, points in zip(bars, batter_points):
    plt.text(bar.get_x() + bar.get_width() / 4 - 0.1, bar.get_height() + 5, str(points), fontsize=10)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
